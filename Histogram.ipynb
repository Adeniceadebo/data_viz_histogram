```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df_income_distribution = pd.read_csv('income_distribution_data.csv')
df_income_distribution 
```

```python
sns.set_theme(style='white')
plt.figure(figsize=(8, 6))
plt.hist(df_income_distribution['Income'], bins=8, color='#9696DB')
plt.title('Income Distribution by Among Survey Respondents', fontsize=14, fontweight='bold')
plt.xlabel('Income', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=12, fontweight='bold')
plt.tight_layout()
sns.despine()
plt.savefig('income_distribution_plot.png', dpi=300)
plt.show()
```
