```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df_income_distribution = pd.read_csv('income_distribution_data.csv')
df_income_distribution 
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8, 6))
plt.hist(df_income_distribution['Income'], bins=30, color='blue', edgecolor='black')
plt.title('Income Distribution by Demographic Factors', fontsize=14, fontweight='bold')
plt.xlabel('Income', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
sns.despine()
plt.savefig('income_distribution_plot.png', dpi=300)
plt.show()
```
